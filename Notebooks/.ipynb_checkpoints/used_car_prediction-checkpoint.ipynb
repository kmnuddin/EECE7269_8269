```python
import pandas as pd
import numpy as np

import statsmodels.formula.api as sm

import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
X_train = pd.read_csv('used car/X_train.csv')
y_train = pd.read_csv('used car/y_train.csv')

X_test = pd.read_csv('used car/X_test.csv')
y_test = pd.read_csv('used car/test_label/y_test.csv')

df = X_train.copy()
df['price'] = y_train['price']
df
```

```python
cols = ['mileage', 'tax', 'mpg', 'engineSize', 'price']

fig, axes = plt.subplots(ncols=5, figsize=(15, 4))


for ax, col in zip(axes, cols):
    sns.kdeplot(data=df, x=col, ax=ax)
    
fig.tight_layout()
cols.remove('price')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 8))


for ax, col in zip(axes.flatten(), cols):
    sns.scatterplot(data=df, x=col, y='price', ax=ax)
    
fig.tight_layout()
```

```python
df['log_mileage'] = df['mileage'].apply(np.log)
sns.scatterplot(data=df, x='log_mileage', y='price')
```

### Forward Selection

```python
lm1 = sm.ols(formula='price ~ log_mileage', data=df).fit()
print(lm1.summary())
```

```python
lm2 = sm.ols(formula='price ~ mileage', data=df).fit()
print(lm2.summary())
```

```python
lm3 = sm.ols(formula='price ~ mileage + mpg', data=df).fit()
print(lm3.summary())
```

```python
lm4 = sm.ols(formula='price ~ mileage + mpg + engineSize', data=df).fit()
print(lm4.summary())
```

#### Introducing categorical variable

```python
df = pd.get_dummies(df, columns=['transmission'], drop_first=True)
df
```

```python
formula = 'price ~ mileage + mpg + engineSize + transmission_Manual + transmission_Other + transmission_Semi-Auto'
lm5 = sm.ols(formula=formula, data=df).fit()
print(lm5.summary())
```
