```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

import statsmodels.api as sm
import statsmodels.formula.api as smf
```

## Load Dataset

```python
df = pd.read_csv('../Datasets/jaybob.csv')
df.head(10)
```

## Visualization

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12, 12))
sns.scatterplot(data=df, x='Odometer', y='Price', ax=ax[0,0])
sns.scatterplot(data=df, x='Age', y='Price', ax=ax[0, 1])

sns.barplot(data=df, x='Pink slip', y='Price', ax=ax[1, 0])
sns.barplot(data=df, x='Sold?', y='Price', ax=ax[1, 1])

fig.tight_layout()
```

## Model 1

```python
model1 = smf.ols(formula='Price ~ Age + Odometer', data=df)
res = model1.fit()
print(res.summary())
```

```python
age = np.linspace(df['Age'].min(), df['Age'].max(), 100)
odometer = np.linspace(df['Odometer'].min(), df['Odometer'].max(), 100)

x = pd.DataFrame({'Age': age, 'Odometer': odometer})
x = sm.add_constant(x)
pred_price = res.predict(x)

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))
sns.scatterplot(data=df, x='Age', y='Price', ax=ax[0])
ax[0].plot(x['Age'], pred_price, color='red', label='Model (R2 = 0.171)')
ax[0].legend()

sns.scatterplot(data=df, x='Odometer', y='Price', ax=ax[1])
ax[1].plot(x['Odometer'], pred_price, color='red', label='Model (R2 = 0.171)')
ax[1].legend()

fig.tight_layout()
```

## Model 2

```python
model2 = smf.ols(formula='Price ~ Age + I(Age**2) + I(1/Odometer)', data=df)
res = model2.fit()
print(res.summary())
```

```python
# Define the range for 'Age' and compute the corresponding 'Age_squared'
age_squared = age ** 2

# Use the mean 'Odometer' for prediction with varying 'Age'
mean_odometer = df['Odometer'].mean()
age_data = pd.DataFrame({'Age': age, 'Age_squared': age_squared, 'Odometer': np.repeat(mean_odometer, 100)})
age_data = sm.add_constant(age_data)  # Add the intercept
pred_price_age = res.predict(age_data)  # Predict the price

# Define the range for 'Odometer' and compute the corresponding 'Odometer_inversed'
odometer_inversed = 1 / odometer

# Use the mean 'Age' and 'Age_squared' for prediction with varying 'Odometer'
mean_age = df['Age'].mean()
mean_age_squared = mean_age ** 2
odometer_data = pd.DataFrame({'Age': np.repeat(mean_age, 100), 'Age_squared': np.repeat(mean_age_squared, 100), 'Odometer': odometer})
odometer_data = sm.add_constant(odometer_data)  # Add the intercept
pred_price_odometer = res.predict(odometer_data)  # Predict the price

# Plotting
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(18, 6))

# Age vs Price
sns.scatterplot(data=df, x='Age', y='Price', ax=ax[0], color='blue')
ax[0].plot(age, pred_price_age, color='red', label='Model Prediction')
ax[0].set_xlabel('Age')
ax[0].set_ylabel('Price')
ax[0].set_title('Price vs Age')
ax[0].legend()

# Age_squared vs Price
sns.scatterplot(data=df, x=df['Age']**2, y='Price', ax=ax[1], color='blue')
ax[1].plot(age_squared, pred_price_age, color='red', label='Model Prediction')
ax[1].set_xlabel('Age Squared')
ax[1].set_ylabel('Price')
ax[1].set_title('Price vs Age Squared')
ax[1].legend()

# Odometer vs Price (Plot with original Odometer values)
sns.scatterplot(data=df, x='Odometer', y='Price', ax=ax[2], color='blue')
ax[2].plot(odometer, pred_price_odometer, color='red', label='Model Prediction')
ax[2].set_xlabel('Odometer')
ax[2].set_ylabel('Price')
ax[2].set_title('Price vs Odometer')
ax[2].legend()

fig.tight_layout()
plt.show()
```
